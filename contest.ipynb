```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('contest/train.csv')
test = pd.read_csv('contest/test.csv')

y = train.num_sold
X = train.drop(['num_sold'], axis=1)
```

```python
X = X.set_index('id')
test
```

```python
from sklearn.preprocessing import OneHotEncoder as OHE

enc = OHE()
cat_cols = ['country', 'store', 'product']
to_be_enc = test[cat_cols]
enc.fit(to_be_enc)
enc_data = enc.transform(to_be_enc).toarray()
enc_df = pd.DataFrame(enc_data, columns=enc.get_feature_names_out(cat_cols))
enc_df
```

```python
test = test.drop(cat_cols, axis=1)
```

```python
test = pd.concat([test, enc_df], axis=1)
```

```python
id = test.id
test = test.set_index('id')
test
```

```python
bruh = X[cat_cols]
enc_X = enc.transform(bruh).toarray()
X_df = pd.DataFrame(enc_X, columns=enc.get_feature_names_out(cat_cols))
X_df
```

```python
X = pd.concat([X.drop(cat_cols, axis=1), X_df], axis=1)
X
```

```python
y
```

```python
X['date'] = pd.to_datetime(X['date'], format='%Y-%m-%d')
X['date'] = X['date'].apply(lambda x: int(x.timestamp()))
test['date'] = pd.to_datetime(test['date'], format='%Y-%m-%d')
test['date'] = test['date'].apply(lambda x: int(x.timestamp()))
from sklearn.model_selection import train_test_split as tts
X_train, X_valid, y_train, y_valid = tts(X, y, random_state=1)
X_train
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error as mae

model2 = XGBRegressor(n_estimators=3000, learning_rate=0.09, n_jobs=12)
model2.fit(X_train, y_train, 
             early_stopping_rounds=10, 
             eval_set=[(X_valid, y_valid)], 
             verbose=True)
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR

model = RFR(n_estimators=128, n_jobs=8, random_state=9)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score as asc
print(mae(y_valid, model.predict(X_valid)))
```
